```python
import pandas as pd
from matplotlib import pyplot as plt

# read CSV file into a DataFrame
data = pd.read_csv('/content/300csv.csv')
```

```python
# Count the occurrences of each severity level
severity_counts = data['Severity'].value_counts()
# Plotting with bars
plt.bar(severity_counts.index, severity_counts.values, color='purple')
plt.xlabel('Severity')
plt.ylabel('Number of Samples')
plt.title('Severity vs Number of Samples')

# Setting x-axis ticks and labels
plt.xticks([0, 1, 2], ['Low', 'Medium', 'High'])

plt.show()
```

```python
# Count the occurrences of each number of casualties
casualty_counts = data['No_of_casualties'].value_counts()

# Plotting bars
plt.bar(casualty_counts.index, casualty_counts.values, color='red')
plt.xlabel('Number of Casualties')
plt.ylabel('Number of Samples')
plt.title('Number of Casualties vs Number of Samples')
plt.xticks(casualty_counts.index)  # Setting x-axis ticks to the unique values in the 'No_of_casualties' column
plt.show()
```

```python
# Plotting histogram
plt.hist(data['Speed'], bins=10, color='green', alpha=0.7)
plt.xlabel('Speed')
plt.ylabel('Frequency')
plt.title('Speed vs Frequency')
plt.show()
```

```python
# Count the occurrences of each weather type
weather_counts = data['Weather'].value_counts()

# Plotting with yellow bars
plt.bar(weather_counts.index, weather_counts.values, color='yellow')
plt.xlabel('Weather')
plt.ylabel('Frequency')
plt.title('Weather vs Frequency')
plt.xticks([1, 2, 3], ['Clear', 'Fog', 'Rainy'])  # Setting x-axis ticks to correspond to weather types
plt.show()
```

```python
# Count the occurrences of each road surface type
road_surface_counts = data['Road_surface'].value_counts().sort_index()

# Plotting with bars
plt.bar(road_surface_counts.index, road_surface_counts.values, color='green')
plt.xlabel('Road Surface')
plt.ylabel('Number of Samples')
plt.title('Road Surface vs Number of Samples')

# Setting x-axis ticks and labels
plt.xticks([1, 2, 3], ['Concrete', 'Gravel', 'Bituminous'])
plt.show()
```

```python
# Count the occurrences of alcohol drunk (Yes or No)
alcohol_counts = data['Alcohol'].value_counts().sort_index()

# Plotting with blue bars
plt.bar(alcohol_counts.index, alcohol_counts.values, color='blue')
plt.xlabel('Alcohol Drunk')
plt.ylabel('Number of Samples')
plt.title('Alcohol Drunk vs Number of Samples')

# Setting x-axis ticks and labels
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python

# Count the occurrences of gradient (Yes or No)
gradient_counts = data['Gradient'].value_counts().sort_index()

# Plotting with orange bars
plt.bar(gradient_counts.index, gradient_counts.values, color='orange')
plt.xlabel('Gradient')
plt.ylabel('Number of Samples')
plt.title('Gradient vs Number of Samples')

# Setting x-axis ticks and labels
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
# Count the occurrences of intersection (Yes or No)
intersection_counts = data['Intersection'].value_counts().sort_index()

# Plotting with bars
plt.bar(intersection_counts.index, intersection_counts.values, color='cyan')
plt.xlabel('Intersection')
plt.ylabel('Number of Samples')
plt.title('Intersection vs Number of Samples')

# Setting x-axis ticks and labels
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
# Count the occurrences of accidents occurred (Yes or No)
accident_counts = data['Accident'].value_counts().sort_index()

# Plotting with bars
plt.bar(accident_counts.index, accident_counts.values, color='magenta')
plt.xlabel('Accident Occurred')
plt.ylabel('Number of Samples')
plt.title('Accident Occurred vs Number of Samples')

# Setting x-axis ticks and labels
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```
