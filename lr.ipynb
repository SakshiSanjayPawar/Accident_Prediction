Using Logistic Regression to predict Accident Prone Hotspots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

#Load data
df=pd.read_csv('/content/300csv.csv')

#Plotting data
plt.scatter(df['Index'],df['Accident'])
plt.show()

#Spliting data into training and testing data set
X_train, X_test, y_train, y_test = train_test_split(df[df.columns[3:11]].to_numpy(),df['Accident'], test_size=0.2)

#Creat Object
lr=LogisticRegression()

#Train data
lr.fit(X_train,y_train)

# Test Model
test=lr.predict(X_test)

#Predict Accuracy of the model
print("Accuracy",lr.score(X_test,y_test)*100)  #accuracy   0.525

#creating dataframe with prediction column
testdf=df1 = df.iloc[int(0.8 * len(df)):]
testdf.insert(12,"Prediction",test)
```
